```python
#import dependencies
import pandas as pd
import numpy as np
import ast
```

```python
#set csv file path
weather_csv = "Resources/rawCitiesWeatherData.csv"
 
# read csv & create dataframe
rawWeatherTable = pd.read_csv(weather_csv)
rawWeatherTable.head()
```

```python
##gather relevant data
#list cities and cityIDs
cityIDs = rawWeatherTable["id"]
cities = rawWeatherTable["name"]

#set empty lists for other columns
countries = []
lats = []
temps = []
humids = []
winds = []
clouds = []

#loop over dataframe and extract data
for i in range(len(rawWeatherTable["id"])):
    countries.append(ast.literal_eval(rawWeatherTable["sys"][i])["country"])
    lats.append(ast.literal_eval(rawWeatherTable["coord"][i])["lat"])
    temps.append(ast.literal_eval(rawWeatherTable["main"][i])["temp"])
    humids.append(ast.literal_eval(rawWeatherTable["main"][i])["humidity"])
    winds.append(ast.literal_eval(rawWeatherTable["wind"][i])["speed"])
    clouds.append(ast.literal_eval(rawWeatherTable["clouds"][i])["all"])

#create data frame
weatherTable = pd.DataFrame({"CityID": cityIDs, 
                              "City": cities, 
                              "Country": countries,
                              "Latitude": lats, 
                              "Temperature (F)": temps, 
                              "Humidity %": humids, 
                              "Wind (MPH)": winds, 
                              "Cloud %": clouds})
#set column order
weatherTable = weatherTable [["CityID", "City", "Country", "Latitude", 
                              "Temperature (F)", "Humidity %", "Wind (MPH)", "Cloud %"]]
weatherTable.head()
```

```python
#export as csv
weatherTable.to_csv("Resources/cleanCitiesWeatherData.csv", index=False)
```
